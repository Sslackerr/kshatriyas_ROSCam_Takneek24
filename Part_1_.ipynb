```python
import cv2
import numpy as np
import os

# Input and output directories
input_dir = r"C:\Users\ksuya\Downloads\06144\06144\rgb" # Directory where the input images are stored
output_dir = r"C:\Users\ksuya\OneDrive - IIT Kanpur\Desktop\New folder (2)" # Directory where the processed images will be saved

# Ensure the output directory exists
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Iterate over each file in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith(".jpg") or filename.endswith(".png"):  # Ensure it's an image file
        image_path = os.path.join(input_dir, filename)
        image = cv2.imread(image_path)

        # Convert the image from BGR to RGB for processing
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Define the lower and upper bounds for the table top (white color) in RGB format
        lower_bound = np.array([200, 200, 200])
        upper_bound = np.array([255, 255, 255])

        # Create a mask for the table top using inRange
        mask = cv2.inRange(image_rgb, lower_bound, upper_bound)

        # Find contours in the mask
        contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Variables to store corner points
        corner1, corner2, corner3, corner4 = None, None, None, None

        # If contours are found, process them
        if contours:
            # Sort contours by area and use the largest one
            largest_contour = max(contours, key=cv2.contourArea)

            # Approximate the contour to a quadrilateral with increased accuracy
            epsilon = 0.02 * cv2.arcLength(largest_contour, True)  # Adjust epsilon for higher precision
            approx = cv2.approxPolyDP(largest_contour, epsilon, True)

            # Ensure the approximation yields exactly 4 points for a quadrilateral
            if len(approx) == 4:  
                # Assign each corner to a variable
                corner1 = tuple(approx[0][0])
                corner2 = tuple(approx[1][0])
                corner3 = tuple(approx[2][0])
                corner4 = tuple(approx[3][0])

                # Draw lines between corners
                quad_image = image_rgb.copy()
                cv2.line(quad_image, corner1, corner2, (0, 0, 255), 3)
                cv2.line(quad_image, corner2, corner3, (0, 0, 255), 3)
                cv2.line(quad_image, corner3, corner4, (0, 0, 255), 3)
                cv2.line(quad_image, corner4, corner1, (0, 0, 255), 3)

                # Highlight the corners on the image
                for corner in [corner1, corner2, corner3, corner4]:
                    cv2.circle(quad_image, corner, 8, (0, 255, 0), -1)  # Draw green circles at corners

            else:
                print(f"Could not approximate a quadrilateral in {filename}.")
                # Fallback to drawing a bounding box if approximation fails
                x, y, w, h = cv2.boundingRect(largest_contour)
                cv2.rectangle(quad_image, (x, y), (x + w, y + h), (0, 0, 255), 3)

        # Save the output image in the output directory
        output_image_path = os.path.join(output_dir, f'processed_{filename}')
        quad_image_bgr = cv2.cvtColor(quad_image, cv2.COLOR_RGB2BGR)  # Convert back to BGR for saving
        cv2.imwrite(output_image_path, quad_image_bgr)

        print(f"Processed image saved at: {output_image_path}")

print("All images processed.")
```

```python
import os
import cv2
import numpy as np

# Paths for depth images and center coordinates
rgb_dir = r"C:\Users\ksuya\Downloads\06144\06144\rgb"  # Directory where the RGB images are stored
depth_dir = r"C:\Users\ksuya\Downloads\06144\06144\depth"  # Directory where the depth images are stored
output_dir = r"C:\Users\ksuya\OneDrive - IIT Kanpur\Desktop\New folder (2)"  # Directory where the processed depth information will be saved
center_txt_path = os.path.join(output_dir, 'center_coordinates.txt')  # File containing center coordinates
depth_output_txt_path = os.path.join(output_dir, 'depth_at_centers.txt')  # File to save depth information

# Ensure the output directory exists
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Load center coordinates from the center_coordinates.txt file
center_coordinates = []
with open(center_txt_path, 'r') as center_file:
    next(center_file)  # Skip the header
    for line in center_file:
        filename, center_x, center_y = line.strip().split(',')
        center_coordinates.append((int(center_x), int(center_y)))

# Sort the RGB and depth images
rgb_images = sorted([f for f in os.listdir(rgb_dir) if f.endswith(".jpg") or f.endswith(".png")])
depth_images = sorted([f for f in os.listdir(depth_dir) if f.endswith(".png")])

# Ensure that the number of center coordinates, RGB images, and depth images are the same


# Variables to track the maximum depth and the corresponding image
max_depth = -1
max_depth_image = None

# Open the depth output text file for writing
with open(depth_output_txt_path, 'w') as depth_file:
    # Write header to the file
    depth_file.write("Filename, Center_X, Center_Y, Depth\n")

    # Process each RGB and depth image pair by index
    for i in range(len(rgb_images)-30):
        rgb_filename = rgb_images[i]
        depth_filename = depth_images[i]
        center_x, center_y = center_coordinates[i]

        # Load the corresponding depth image
        depth_image_path = os.path.join(depth_dir, depth_filename)
        depth_image = cv2.imread(depth_image_path, cv2.IMREAD_UNCHANGED)

        if depth_image is not None:
            # Retrieve the depth value at the center
            depth_value = depth_image[center_y, center_x]

            # Write the depth value to the file for each image
            depth_file.write(f'{rgb_filename}, {center_x}, {center_y}, {depth_value}\n')

            # Check if this is the maximum depth and update if necessary
            if depth_value > max_depth:
                max_depth = depth_value
                max_depth_image = rgb_filename
        else:
            print(f"Error loading depth image: {depth_filename}")

# Output the image with the maximum depth
if max_depth_image:
    print(f"The image with the maximum depth is {max_depth_image} with a depth of {max_depth}")
else:
    print("No valid depth data found.")

print(f"Depth values saved at: {depth_output_txt_path}")
```
